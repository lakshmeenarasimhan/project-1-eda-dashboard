```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/zomato.csv", encoding="latin-1")
df.shape
```

```python
drop_cols = ['url', 'phone', 'menu_item', 'dish_liked', 'reviews_list']
df.drop(columns = drop_cols,inplace = True)
print(df.shape)
df.columns.tolist()
```

```python
df['rate'].unique()
# Doing this to know what different types of values are there in the column
```

```python
#changing the dataype and replacing error values with NaN
df['rate'] = df['rate'].str.replace('/5','',regex = False)
df['rate'] = df['rate'].str.strip()
df['rate'] = df['rate'].replace('NEW',np.nan)
df['rate'] = df['rate'].replace('-',np.nan)
df['rate'] = df['rate'].astype(float)

print(df['rate'].dtype)
print(df['rate'].unique())
```

```python
df['approx_cost(for two people)'].unique()
```

```python
#changing the dataype and replacing error values with NaN
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','',regex = False)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)

print(df['approx_cost(for two people)'].dtype)
print(df['approx_cost(for two people)'].unique())
```

```python
df.isnull().sum()
```

```python
df  = df.dropna(subset = ['rate','approx_cost(for two people)','rest_type', 'cuisines', 'location'])
df.shape
```

```python
df.duplicated().sum()
```

```python
#removing dupliactes from the dataset
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df.shape
```

```python
# final summary
print("Shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())
print("\nData types:")
print(df.dtypes)
```

```python
df.to_csv("../data/cleaned/zomato_cleaned.csv", index=False)
print("Saved successfully!")
```
